```python
import torch
import torch.nn as nn
```

```python
input_dim = 4
hidden_dim = 6
batch_size = 3
seq_len = 5
```

```python
inp = torch.rand(size = (batch_size, seq_len, input_dim), dtype=torch.float)
inp.shape
```

```python
# hidden_state (= short-term memory)
hidden_state = torch.zeros(1,batch_size, hidden_dim, dtype=torch.float)

# cell_state (= long-term memory)
cell_state = torch.zeros(1, batch_size, hidden_dim, dtype=torch.float)

# hidden
hidden = (hidden_state, cell_state)

# LSTM 모델을 생성
lstm_layer = nn.LSTM(input_dim, hidden_dim, batch_first=True)
```

```python
hidden_state
```

```python
out, hidden = lstm_layer(inp, hidden)
hidden[0]
```
