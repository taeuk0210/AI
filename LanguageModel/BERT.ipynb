# Pre-training of Deep Bidirectional Transformers

https://arxiv.org/pdf/1810.04805.pdf

BERT는 Bidirectional Encoder Representations from Transformers의 약자로 unlabeled text로 부터 jointly conditioning을 통해 양방향 representation을 학습하여 pre-training 된 BERT는 output layer의 추가만으로 다양한 task에서 좋은 성능을 보임  

보통 language model은 unidirectional하기 때문에 이전의 방법들은 pre-trained representation의 영향을 제한시키며 모델 선택에 있어서도 제한적임

image.png

BERT 는 pre-training objective 로 masked language model 을 사용함으로써 unidirectional 한 한계점들을 완화시킴.
(masked language model : random 하게 input 의 token 몇 개를 mask 하여 mask 된 token 이 뭔지 예측하는 것)
기존 언어모델과 달리 MLM objective 는 양쪽의 context의 representation을 합칠 수 있음

Contributions :  
1. language representation에서 bidirectional pre-training의 중요성을 보여줌
2. pre-trained representation이 task-specific한 여러 과정들을 줄여줌
3. 11개의 nlp tasks에서 좋은 성능을 발휘함

multi-layer bidirectioinal transformer encoder를 이용하며 framework는 두 stage로 나뉜다.  
**step 1** : unlabeled data로 pre-training,
two unsupervised task를 사용  
 - Masked LM : tokens 의 15%를 random하게 masking, (80% : mask, 10% : random token, 10% : unchanged) 
 - Next Sentence Prediction : 두 문장간의 관계를 학습하기 위함 

**step 2** : tasks에 맞는 labeled data로 fine-tuning
