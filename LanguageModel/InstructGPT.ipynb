# Training language models to follow instrcutions with human feedback

https://arxiv.org/pdf/2203.02155.pdf

large LM 은 잘못된 정보나, 유해한 정보 또는 유저에게 도움을 주지 못하는 정보를 생성할 수 있다.  
여기에서는 유저의 의도에 맞게 LM 을 aligning 하는 방법을 소개  

사람의 선호도를 reward로 하여 fine-tuning 즉 사람이 더욱 선호하는 답을 만들어 낼 수 있도록 함  

main finding :
 - Labelers GPT-3보다 InstructGPT의 결과를 선호
 - GPT-3보다 진실성에서 향상됨
 - GPT-3보다 유해성에서 약간 향상되나 편향은 그렇지 않음
 -  performance regression 최소화
 - 일반화 능력 향상

**step 1**
 - prompt dataset에서 prompt 샘플링
 - labeler가 prompt에 대한 결과를 생성 
 - pre-trained GPT-3을 fine-tuning하여 SFT를 얻음  

**step 2**
 - prompt와 그에 따른 여러 모델의 결과물들을 샘플링
 - labeler가 선호도 순위를 라벨링
 - 이 data를 이용하여 reward model 학습  

**step 3**
 - step 1의 fine-tuning된 모델을 step 2의 reward model을 이용하여 강화학습을 통해 한번 더 fine-tuning
