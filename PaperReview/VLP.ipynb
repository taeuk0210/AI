## Abstract 

1. fine-tuning for vision-language generation or understanding
2. shared multi-layer transformer network
3. bidirectional and seq2seq masked vision-language prediction
4. specific self-attention masks for shared transformer

## Introduction

single, unified model(universally applicable, fine-tuning)을 pre-train하는 것이 매우 어려움  

대부분의 pre-trained model은 understanding task 만으로 설계되거나 따로 multimodality-specific encoder, decoder을 학습시켜 사용하는 hybrid model이다.

pre-training a unified representation for both encoding and decoding, eliminating the discrepancy(pre-train시 decoder는 학습 x) 방법을 제시

## Vision-Language Pre-training

image.png

**Pre-training objectives**  

bidirectional objective : 
> 15% of the input text tokens are replaced :  
[MASK] token(80%), random token(10%), original token(10%) > predict the masked tokens

seq2seq objective :  
> source segment : bidirectional LM  
> target segment : left-to-right LM > predict masked tokens  

위의 pre-training objective는 self-attention mask를 달리하여 수행할 수 있음  

나머지 구조는 transformer 그대로 사용

image.png

**Fine-Tuning for Downstream Tasks**  

Image Captioning :  
> seq2seq objective로 fine-tuning  
> encode the image region > generate description  

Visual Question Answering :  
> multi-label classification
