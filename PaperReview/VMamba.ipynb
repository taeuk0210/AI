# **VMamba : Visual State Space Model**

## **Introduction**



computer vision분야에서 CNN, ViT는 현재까지 가장 많이 사용되는 카테고리로 다양한 visual task에서 이용되고 있음  

ViT는 global receptive field and dynamic weights facilitated by the attention mechanism으로 CNN보다 높은 성능을 가짐  

그러나 attention mechanism은 image size에 대한 quadratic complexity를 가져 downstream dense prediction task시 계산 비용이 크다  

저자들의 main contributions :  
1. propose VMamba with global receptive fields and dynamic weights for visual representation learning  
2. 1-D array scanning 과 2-D plain traversing의 차이를 해소하기 위한 Cross-Scan Module을 소개  
3. 다른 특별한 추가 없이 VMamba는 다양한 visual task에서 잠재력을 가짐
